## Obtener el valor K

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
##kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','yellow','blue','cyan']
asignar=[]
for row in labels:
    asignar.append(colores[row])
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], c=asignar,s=50)
ax.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
```

```python
# Getting the values and plotting it
f1 = dataset2['CANTIDAD_PRODUCTOS'].values
f2 = dataset2['CANTIDAD_ENTIDADES'].values
 
plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000,)
##plt.legend()
plt.xlabel('CANTIDAD_PRODUCTOS')
plt.ylabel('CANTIDAD_ENTIDADES')
plt.show()
```
